```python
#Housing veri setini kullanıyoruz.
#Amaç bir değişkendeki değişiklerin ev fiyatını nasıl etkilediğini
#görmemizi sağlayacak katsayılar bulmak.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline 

df = pd.read_csv('Housing.csv')
df.info()
df.head()
```

```python
df.describe().T
```

```python
#Hedefimiz ev fiyatları olduğu için median_house_value değişkeninin dağılımına bakıp ayıracağız.
plt.figure(figsize=(6,4))
X = df.drop(['median_house_value'], axis=1)
y = df['median_house_value']
sns.distplot(y)
plt.show()
```

```python
X = pd.concat([X, pd.get_dummies(X.ocean_proximity)], axis=1)
X = X.drop(['ocean_proximity'],axis=1)
#İçinde eksikler olan toplam yatak odası değişkenlerinin içini o değişkenin medyan değeriyle dolduruyoruz. 
X['total_bedrooms'] = X['total_bedrooms'].fillna(X['total_bedrooms'].median())
```

```python
#Eğitim ve test veri setlerine ayırdık veri setimizi.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = X[['longitude', 'latitude', 'housing_median_age', 'households', 'median_income',
       '<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
#Karar ağacı modelini oluşturup eğitim verisine 'fit2 ediyoruz.
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print("R score: {0}".format(round(lm.score(X_train, y_train),2)))
```

```python
print("Intercept: {0}".format(round(lm.intercept_),))
```

```python
pd.DataFrame({'feature':X.columns, 'coef':lm.coef_}) #coefficient
```

```python
#Gerçek değerler ile tahmin değerleri arasındaki ilişkinin grafiği.
plt.figure(figsize=(12,8))
plt.scatter(y_test, lm.predict(X_test), color='r')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Comparison")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

mae = mean_absolute_error(lm.predict(X_test), y_test)
mse = mean_squared_error(lm.predict(X_test), y_test)
rmse = np.sqrt(mse)

print('Mean Absolute Error (MAE): %.2f' % mae)
print('Mean Squared Error (MSE): %.2f' % mse)
print('Root Mean Squared Error (RMSE): %.2f' % rmse)
```

```python
#Scikit-learn ile yapılan işin statsmodels.api ile yapılmış hali
import statsmodels.api as sm

model = sm.OLS(y_train, X_train).fit()
print('Mean Absolute Error (MAE): %.2f' % mae)
print('Mean Squared Error (MSE): %.2f' % mse)
print('Root Mean Squared Error (RMSE): %.2f' % rmse)

model.summary()
```
